# Affinity calculation

### This codes calculates Drug-Gene affinities (pKd) 

```python
import pandas as pd
import numpy as np
from DeepPurpose import utils
from DeepPurpose import DTI as models
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#opening the files
gene_file  = '..\gene_full_pI.txt'
drug_file  = '..\SIDER-SMILES.csv'
#gene_file  = 'toy_genes3.txt'
#drug_file  = 'toy_drugs.csv'

genes = pd.read_csv(gene_file, sep=',')
drugs = pd.read_csv(drug_file, sep=',')
```

```python
#list of 14 pretrained models in DeepPurpose
model_names = ['CNN_CNN_DAVIS','CNN_CNN_BindingDB','Morgan_CNN_BindingDB','Morgan_AAC_DAVIS',
               'Morgan_CNN_DAVIS','MPNN_CNN_BindingDB','MPNN_CNN_KIBA','MPNN_CNN_DAVIS','Morgan_AAC_KIBA',
               'Transformer_CNN_BindingDB','Daylight_AAC_DAVIS','Daylight_AAC_KIBA','Daylight_AAC_BindingDB',
               'Morgan_AAC_BindingDB']
```

```python
#downloads and extracts the models
for model in model_names:
   # utils.download_pretrained_model(model) 
```

```python
#Perdicts the DTI using pretrained models
def DTI_pred(pre_model, drug, target):
    model_info = pre_model.split('_')  #Seperates data and encodings 
    model = models.model_pretrained(model = pre_model)
    X_pred = utils.data_process(X_drug = [drug], X_target = [target], y = [0],
                                drug_encoding = model_info[0], target_encoding = model_info[1], 
                                split_method='no_split')
    y_pred = model.predict(X_pred)  #Perdicts affinity
    return str(round(y_pred[0],2))
```

```python
%%time

dictinary_list = []
for d_indx in range(len(drugs)):
    drug = drugs['SMILES'][d_indx]
    #print(d_indx)
    for g_indx in range(len(genes)):
        target = genes['Sequence'][g_indx]

        row =  {'DrugID': drugs['PubChemID'][d_indx],
                'GeneID': genes['gene_id'][g_indx],
                'pI': genes['pI'][g_indx],
                'SMILES': drug,
                'Sequence': target}    
        
        dictinary_list.append(row)


affinity = pd.DataFrame.from_dict(dictinary_list)

affinity['MPNN_CNN_DAVIS'] = affinity.apply(lambda x: DTI_pred('MPNN_CNN_DAVIS',x['SMILES'], x['Sequence']),axis=1)
```
