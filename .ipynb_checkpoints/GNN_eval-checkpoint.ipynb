```python
import dgl
import torch
import numpy as np
import pandas as pd
import networkx as nx
from tqdm.notebook import tqdm
from urllib.request import urlopen
```

```python
# Eval combinations
eval_df = pd.read_csv('../data/TWOSIDE-evaluation-PSE-964.csv', sep=',')
eval_df
```

```python
drugs_list = eval_df['Drug1'].unique().tolist()
for drug in eval_df['Drug2'].unique().tolist():
    if drug not in drugs_list:
        drugs_list.append(drug)

len(drugs_list)
```

```python
%%time

f = open('../data/eval_drugs_964.tsv', 'a')
f.write('Drug_name\tSMILES\n')

for drug in tqdm(drugs_list):
    drug_name = drug
    
    if ' ' in drug:
        drug = drug.replace(' ', '%20')
        
    try:
        url = 'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/'+drug+'/property/CanonicalSMILES/TXT'
        res = urlopen(url)
        smiles = str(res.read())[2:-3]
        row = drug_name + '\t' + smiles + '\n'
        f.write(row)
    except:
        #row = drug_name + '\t' + '-\n'
        #f.write(row)
        pass

f.close()
```
